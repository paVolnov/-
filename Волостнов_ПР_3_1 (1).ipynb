```python
from google.colab import files
import pandas as pd


uploaded = files.upload()


file_name = "Global YouTube Statistics.csv"


data = pd.read_csv(file_name, encoding='cp1251')
```

```python
data.head(5)
```

```python
shape = data.shape
print("Розмір датасету: ", shape)
```

```python

data = data.apply(lambda x: pd.to_numeric(x, errors='coerce'))


missing_values = data.isna().sum()
print("Кількість пропусків у кожному стовпці:")
print(missing_values)
```

```python

data = data.fillna(data.mean())


data = data.astype(float)
```

```python
missing_values = data.isna().sum()
print("Кількість пропусків у кожному стовпці після заміни:")
print(missing_values)
```

```python
unique_countries = data['Country'].nunique()
print("Кількість унікальних країн:", unique_countries)

```

```python
import matplotlib.pyplot as plt
```

```python
data['video views'].plot(kind='hist', bins=20, edgecolor='k')
plt.xlabel('Перегляди')
plt.ylabel('Частота')
plt.title('Розподіл переглядів')
plt.show()
```

```python

max_views = data['video views'].max()
print("Максимальна кількість переглядів:", max_views)


min_views = data['video views'].min()
print("Мінімальна кількість переглядів:", min_views)


mean_views = data['video views'].mean()
print("Середня кількість переглядів:", mean_views)
```

```python

max_uploads = data['uploads'].max()
video_with_max_uploads = data[data['uploads'] == max_uploads]
print("Відео з найбільшою кількістю завантажень:")
print(video_with_max_uploads['Title'].values[0])


min_uploads = data['uploads'].min()
video_with_min_uploads = data[data['uploads'] == min_uploads]
print("Відео з найменшою кількістю завантажень:")
print(video_with_min_uploads['Title'].values[0])
```
